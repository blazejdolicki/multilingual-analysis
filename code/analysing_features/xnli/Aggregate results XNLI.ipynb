```python
import os, pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
task="xnli"
```

```python
results = {"train_lang":[],"test_lang":[],"acc":[]}
for file in os.listdir("results/"):
    if "dev_results" in file:
        train_lang = file.split("_")[2][:2]
        with open(f"results/{file}") as f:
            lines = f.read().strip().split("\n")
            f.close()

        for i,line in enumerate(lines):
            if i==0:
                continue
            test_lang = line.split("=")[0]
            acc = line.split("=")[1]
            if test_lang != "total":
                results["test_lang"].append(test_lang)
                results["train_lang"].append(train_lang)
                results["acc"].append(acc)

results_df = pd.DataFrame.from_dict(results)
```

```python
results_df.to_csv(f"{task}_lang_pair_scores.csv",index=False)
```
