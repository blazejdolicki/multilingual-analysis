```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
enc.fit(xtreme_langs[lang_feats])
```

```python
list(zip(enc.transform([xtreme_langs.loc["en",lang_feats]])[0],enc.transform([xtreme_langs.loc["ru",lang_feats]])[0]))
```

```python
from sklearn.metrics import pairwise
from sklearn.metrics import accuracy_score
import numpy as np
num_langs = xtreme_langs.shape[0]
lang_sim = np.zeros((num_langs,num_langs))
for i,lang1 in enumerate(xtreme_langs.index):
    for j,lang2 in enumerate(xtreme_langs.index):
    #     sim = pairwise.cosine_similarity(enc.transform([xtreme_langs.loc[lang1,lang_feats]]),enc.transform([xtreme_langs.loc[lang2,lang_feats]]))
        sim = accuracy_score(enc.transform([xtreme_langs.loc[lang1,lang_feats]])[0],enc.transform([xtreme_langs.loc[lang2,lang_feats]])[0])
        lang_sim[i][j] = sim
```

```python
similarities_df = pd.DataFrame(data=lang_sim,columns=xtreme_langs.index,index=xtreme_langs.index)
```

```python
from collections import defaultdict
import numpy as np
letter_missing = defaultdict(list)
letters = []
for col in wals.columns[10:]:
    letter = col.split()[0][-1]
    letter_missing[letter].append(wals.shape[0]-wals[col].count())

for letter in letter_missing:
    letters.append([letter,int(np.mean(letter_missing[letter])),len(letter_missing[letter])])
```

```python
df = pd.DataFrame(data=letters,columns=['letter','avg NaNs','frequency'])
```
