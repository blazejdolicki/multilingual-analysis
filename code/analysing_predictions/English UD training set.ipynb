```python
import pandas as pd

pd.options.display.max_rows = 100000
pd.set_option('display.max_colwidth', -1)
```

```python
with open("train-en.tsv",encoding="utf-8") as f:
    lines = f.read().split("\n")

all_sentences = []
all_tags = []
sentence = ""
sentene_tags = ""
i = 0
for line in lines:
#     if a sentence ended
    if line=="":
        i += 1
        all_sentences.append(sentence.strip())
        all_tags.append(sentene_tags.strip())
        sentence =""
        sentene_tags = ""
    else:
        word = line.split("\t")[0]
        tag = line.split("\t")[1]
        
        sentence += " "+word
        sentene_tags += " "+tag
```

```python
ud_pos = pd.DataFrame(data={"sentences":all_sentences,"ground_truth":all_tags})
```
