```python
import pandas as pd

pd.options.display.max_rows = 100000
pd.set_option('display.max_colwidth', -1)
```

```python
with open("dev-nl.tsv",encoding="utf-8") as f:
    lines = f.read().split("\n")

all_sentences = []
all_tags = []
sentence = ""
sentene_tags = ""
i = 0
for line in lines:
#     if a sentence ended
    if line=="":
        i += 1
        all_sentences.append(sentence.strip())
        all_tags.append(sentene_tags.strip())
        sentence =""
        sentene_tags = ""
    else:
        word = line.split("\t")[0]
        tag = line.split("\t")[1]
        
        sentence += " "+word
        sentene_tags += " "+tag
```

```python
ud_pos = pd.DataFrame(data={"sentences":all_sentences,"ground_truth":all_tags})
```

```python
ud_pos[ud_pos.ground_truth.str.contains("PROPN")].head(100)
```

```python
total_tags = 0
incorrect_tags = 0
mistakes_df = {"actual":[],"token":[],"sentence":[],"actual_tokens":[]}
for i in range(len(all_tags)):
    actual_sent = all_tags[i].split()
    for j in range(len(actual_sent)):
        if actual_sent[j]=="PROPN":
            mistakes_df["actual"].append(actual_sent[j])
            mistakes_df["token"].append(all_sentences[i].split()[j])
            mistakes_df["sentence"].append(all_sentences[i])
            mistakes_df["actual_tokens"].append(" ".join(actual_sent))
            
            incorrect_tags += 1
        total_tags += 1
#     if incorrect_tags==5:
#         print("break")
#         break
```

```python
mistakes_df = pd.DataFrame.from_dict(mistakes_df)
```
