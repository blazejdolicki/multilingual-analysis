```python
iso_codes = pd.read_excel("xtreme_iso_codes.xlsx")
```

```python
set(iso_codes["Language name"])-set(wals["Name"])
```

```python
xtreme_langs = wals[wals["Name"].isin(list(iso_codes["Language name"].values)+["Greek (Modern)","Arabic (Modern Standard)","Hebrew (Modern)","Mandarin"])]
```

```python
xtreme_langs = xtreme_langs.set_index("iso_code")
```

```python
num_features = xtreme_langs.loc[:,"20A Fusion of Selected Inflectional Formatives":].count(axis=1)
```

```python
xtreme_langs.insert(9, '#feats', num_features)
```

```python
lang_feats = xtreme_langs.columns[10:]
```

```python
xtreme_langs.to_excel("xtreme_langs_wals.xlsx")
```

```python
xtreme_langs.sort_values(by="#feats")
```

```python
xtreme_langs[lang_feats] = xtreme_langs[lang_feats].fillna("Missing")
```

```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
enc.fit(xtreme_langs[lang_feats])
```

```python
list(zip(enc.transform([xtreme_langs.loc["eng",lang_feats]])[0],enc.transform([xtreme_langs.loc["rus",lang_feats]])[0]))
```

```python
from sklearn.metrics import pairwise
from sklearn.metrics import accuracy_score
import numpy as np
similarities = {"lang":[],"sim":[]}
num_langs = xtreme_langs.shape[0]
lang_sim = np.zeros((num_langs,num_langs))
for i,lang1 in enumerate(xtreme_langs.index):
    for j,lang2 in enumerate(xtreme_langs.index):
    #     sim = pairwise.cosine_similarity(enc.transform([xtreme_langs.loc[lang1,lang_feats]]),enc.transform([xtreme_langs.loc[lang2,lang_feats]]))
        sim = accuracy_score(enc.transform([xtreme_langs.loc[lang1,lang_feats]])[0],enc.transform([xtreme_langs.loc[lang2,lang_feats]])[0])
        lang_sim[i][j] = sim

similarities_df = pd.DataFrame(data=lang_sim,columns=xtreme_langs.Name.values,index=xtreme_langs.Name.values)
```

```python
similarities_df["German"].sort_values(ascending=False)
```

```python
similarities_df.sort_values(by="sim",ascending=False)
```
